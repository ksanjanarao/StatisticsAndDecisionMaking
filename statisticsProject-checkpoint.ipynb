1. For the suburb Altona, it is postulated that a typical property sells for 8,00,000. Use the data at hand to test this assumption.Is the typical property price really $800,000 or has it increased? Use a significance level of 5%.

```python
pip install pandas scipy
```

```python
from scipy import stats
```

Filter the dataset for Altona

```python
df['Suburb'] = df['Suburb'].astype(str)
```

```python
altona_data = df[df['Suburb'] == 'Altona']
altona_data
```

In Case if we need to calculate the property price only for Altona.

```python
altona_price1 = altona_data['Price']
altona_price1
```

```python
# Define the assumed typical price
assumed_price = 800000

# Perform a one-sample t-test
t_statistic, p_value = stats.ttest_1samp(altona_price1, assumed_price)

# Calculate the mean price
mean_price = altona_price1.mean()

# Significance level
alpha = 0.05

# Output results
print(f'Mean Price in Altona: ${mean_price:.2f}')
print(f'T-statistic: {t_statistic:.4f}')
print(f'P-value: {p_value:.4f}')


# Determine if we reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis: The typical property price has significantly changed from $800,000.")
else:
 
   print("Fail to reject the null hypothesis: The typical property price is not significantly different from $800,000.")
```

2. For the year 2016, is there any difference in prices of properties sold in the summer months vs winter months? Consider months from October till March as winter months and the rest as summer months. Use a significance level of 5%.

```python
df['Date'] = df['Date'].astype(str)
```

```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format="%d/%m/%Y")
df['Date']
```

```python
# Filter for the year 2016

data_2016 = df[df['Date'].dt.year == 2016]

# Create a new column for the month

data_2016['Month'] = data_2016['Date'].dt.month

# Define winter and summer months

winter_months = [10, 11, 12, 1, 2, 3]  
# October to March

summer_months = [4, 5, 6, 7, 8, 9]    
# April to September

# Separate the prices based on the season

winter_prices = data_2016[data_2016['Month'].isin(winter_months)]['Price']

summer_prices = data_2016[data_2016['Month'].isin(summer_months)]['Price']

# Perform an independent two-sample t-test
t_statistic, p_value = stats.ttest_ind(winter_prices, summer_prices, equal_var=False)

# Calculate means

mean_winter_price = winter_prices.mean()

mean_summer_price = summer_prices.mean()

# Significance level
alpha = 0.05

# Output results
print(f'Mean Winter Price: ${mean_winter_price:.2f}')

print(f'Mean Summer Price: ${mean_summer_price:.2f}')
print(f'T-statistic: {t_statistic:.4f}')
print(f'P-value: {p_value:.4f}')

# Determine if we reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in property prices between summer and winter months.")
else:
 
  print("Fail to reject the null hypothesis: There is no significant difference in property prices between summer and winter months.")
```

3. For the suburb Abbotsford, what is the probability that out of 10 properties sold, 3 will not have car parking? Use the column car in the dataset. Round off your answer to 3 decimal places.

```python
from scipy.stats import binom
```

```python
# Filter for the suburb Abbotsford
abbotsford_data = df[df['Suburb'] == 'Abbotsford']
abbotsford_data
```

```python
# Calculate the probability of a property not having car parking

total_properties = len(abbotsford_data)
properties_without_car = abbotsford_data[abbotsford_data['Car'] == 0]

probability_no_car = len(properties_without_car) / total_properties
probability_no_car
```

```python
# Define the parameters for the binomial distribution

n = 10 # total number of properties

k = 3  # number of properties without car parking

# Calculate the probability of exactly k properties without car parking
probability_exactly_k = binom.pmf(k, n, probability_no_car)

# Round the result to 3 decimal places
rounded_probability = round(probability_exactly_k, 3)

# Output the result
print(f'The probability that out of 10 properties sold in Abbotsford, 3 will not have car parking is: {rounded_probability}')
```

4. In the suburb Abbotsford, what are the chances of finding a property with 3 rooms? Round your answer to 3 decimal places.
